```python
met ='gucmetM1.b1'
start = '2021-12-10'
end = '2022-03-21' 
```

```python
# Download SAIL met data
if not os.path.exists(outpath+'/met'):
    print('Downloading Data...')
    act.discovery.download_data(username, token, met,
                                start, end, output=outpath+'/met')
else: print(f'Data already downloaded. Located here: {outpath}/met')
```

```python
met_ds=act.io.armfiles.read_netcdf(os.path.join(outpath,f'met/*'))
```

```python
WindDisplay = act.plotting.WindRoseDisplay(met_ds.sel(time=slice('2021-12-23','2022-01-01')), figsize=(8, 10),subplot_shape=(1,))
WindDisplay.plot('wdir_vec_mean', 'wspd_arith_mean',
                 spd_bins=np.linspace(0, 10, 5), num_dirs=30,
                 tick_interval=2, subplot_index=(0,),
                 set_title='SAIL Wind Speed - Dec 23-Jan 1')
```
