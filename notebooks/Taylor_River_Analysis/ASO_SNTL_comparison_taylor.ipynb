### Get ASO files, load up and compute basin mean value

```python
ASO_swe_files = glob.glob('/storage/dlhogan/sos/data/ASO/ASO_50M_SWE.001_TR/*.tif')
```

```python
# skipping the CB area flight on 20180330 because there was another flight on march 30 over our specific basin
dates = [dt.date(2018,3,30),dt.date(2019,4,8),dt.date(2019,6,9),dt.date(2022,5,25),dt.date(2022,4,21)]

# create empty dictionaries to fill
aso_swe_mean_dict = {}
aso_swe_median_dict = {}
aso_swe_25th_dict = {}
aso_swe_75th_dict = {}
aso_swe_date_dict = dict(zip(ASO_swe_files, dates))

for i,file in enumerate(ASO_swe_files):
    src_swe = rioxr.open_rasterio(file).squeeze()
    src_er_swe_ma = src_swe.where(src_swe>=0)
    
    # Save mean and standard deviation in centimeters
    aso_swe_mean_dict[file] = src_er_swe_ma.mean().values * 100 # convert to centimeters
    aso_swe_median_dict[file] = src_er_swe_ma.median().values * 100 # convert to centimeters
    aso_swe_25th_dict[file] = src_er_swe_ma.quantile(0.25).values * 100 # convert to centimeters
    aso_swe_75th_dict[file] = src_er_swe_ma.quantile(0.75).values * 100 # convert to centimeters
    del src_swe,src_er_swe_ma
```

```python
aso_stats_df = pd.DataFrame([aso_swe_date_dict.values(),aso_swe_mean_dict.values(), aso_swe_median_dict.values(), aso_swe_25th_dict.values(), aso_swe_75th_dict.values()]).T
aso_stats_df = aso_stats_df.rename(columns={0:'date',1:'mean', 2:'median', 3:'25th_quartile', 4:'75th_quartile'})
aso_stats_df = aso_stats_df.set_index('date')
```

```python
# Create a plot for a time series of each SNOTEL site from 2016 to 2022
er_sntl_2016_2022 = er_sntl_data.sel(Date=slice('2017-10-01','2022-09-30'))

fig, axs = plt.subplots(nrows=2, ncols=2,figsize=(15,10), tight_layout=True)
axs=axs.ravel()
for i,loc in enumerate(er_sntl_2016_2022.Location):
    ax = axs[i]
    er_sntl_2016_2022.sel(Location=loc).WTEQ.plot(ax=ax)
    
    
    # ax.scatter(aso_stats_df.index, aso_stats_df['mean'], label='ASO mean SWE', color='red',)
    ax.errorbar(aso_stats_df.index, aso_stats_df['median'],yerr=[aso_stats_df['25th_quartile'],aso_stats_df['75th_quartile']], 
                fmt='o', 
                color='black',
                ecolor='gray', 
                elinewidth=3, 
                capsize=0,
                label='ASO mean SWE IQR');
    
    ax.set_title(f'{loc.values} SWE')
    ax.set_ylim(0,130)
    if i == 0:
        ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
er_sntl_2016_2022.sel(Location=['ParkCone_680:CO:SNTL', 'UpperTaylor_737:CO:SNTL']).WTEQ.mean(dim='Location').plot(label='Park Cone/Upper Taylor Mean')

ax.errorbar(aso_stats_df.index, aso_stats_df['median'],yerr=[aso_stats_df['25th_quartile'],aso_stats_df['75th_quartile']], 
                fmt='o', 
                color='black',
                ecolor='gray', 
                elinewidth=3, 
                capsize=0,
                label='ASO median SWE IQR');

ax.set_ylim(0,130)
ax.set_title('Park Cone/Upper Taylor Mean with ASO Median data')
ax.legend()
```

```python
park_cone_upper_taylor_mean_aso_overlap = er_sntl_2016_2022.sel(Date=dates,Location=['ParkCone_680:CO:SNTL','UpperTaylor_737:CO:SNTL']).WTEQ.mean(dim='Location')
park_cone_aso_overlap = er_sntl_2016_2022.sel(Date=dates,Location=['ParkCone_680:CO:SNTL']).WTEQ.mean(dim='Location')
upper_taylor_aso_overlap = er_sntl_2016_2022.sel(Date=dates,Location=['UpperTaylor_737:CO:SNTL']).WTEQ.mean(dim='Location')


plt.plot(dates,(park_cone_aso_overlap-aso_stats_df['median']),label='Park Cone')
plt.plot(dates,(park_cone_upper_taylor_mean_aso_overlap-aso_stats_df['median']),label='Park Cone/Upper Taylor Mean')
plt.plot(dates, (upper_taylor_aso_overlap-aso_stats_df['median']), label='Upper Taylor')
plt.xlabel('Date')
plt.ylabel('SNTL-ASO median SWE Difference (cm)')
plt.title('SNTL - ASO median SWE Differences')
plt.legend()


labels = ['Park Cone/Upper Taylor','Park Cone','Upper Taylor']
for i,site in enumerate([park_cone_upper_taylor_mean_aso_overlap,park_cone_aso_overlap,upper_taylor_aso_overlap]):
    rmse =round(np.sqrt((((site-aso_stats_df['median'])**2/len(dates)).values).mean()),2)
    print(f'{labels[i]} - Mean RMSE = {rmse} cm')
```

### Butte-Schofield Mean is the closest to mean ASO flight values mean values, but Butte and Upper Taylor are best with respect to the median
